```python
url_template="http://www.indeed.com/jobs?q=data_scientist&1=USA&start={}"
max_results=100
Linkdf=[]

for start in range(0, max_results, 10):
    url=url_template.format(start)
    html=requests.get(url)
    soup=BeautifulSoup(html.content,'html.parser', from_encoding="utf-8")
    
    #for each in soup.find_all(a_="href"):
    page_links=soup.find_all('a',{'href':re.compile("/rc/")})
    for items in page_links:
        Linkdf.append(items['href'])
        print(Linkdf)
    
```

```python
def parse(url):
    html=requests.get(url)
    soup=BeautifulSoup(html.content,'html.parser', from_encoding="utf-8")
    df=pd.DataFrame(columns=["Title",  "Salary","Location", "Company", "Description"])
    
    for each in soup.find_all(class_="jobsearch-DesktopStickyContainer"):
        try:
            title=each.find(class_='jobsearch-JobInfoHeader-title').text
        except:
            title='None'
            
        try:
            salary=each.find(class_='jobsearch-JobMetadataHeader-item').text
        except:
            salary='None'
    
    
    for each in soup.find_all(class_="jobsearch-CompanyInfoWithoutHeaderImage"):
        try:
            location=each.find('div', class_='icl-u-lg-mr--sm icl-u-xs-mr--xs').find_next('div').find_next('div').text
            #location=each.find('div', class_='icl-u-lg-mr--sm icl-u-xs-mr--xs').find_next('div').find_next('div').text
            #location=each.find('title').text
            #location=each.find('div',class_='jobsearch-CompanyInfoWithoutHeaderImage').find_next('div').text
            #location=each.find(class_='icl-u-xs-mt--xs  jobsearch-JobInfoHeader-companyLocation jobsearch-DesktopStickyContainer--companylocation').text
            #location=each.find(class_='icl-u-xs-mt--xs').find_next('a').find_next('div').text
            #location=each.find(class_='icl-u-xs-mt--xs  jobsearch-JobInfoHeader-companyLocation jobsearch-DesktopStickyContainer--companylocation').text
            #location=each.find(class_='jobsearch-DesktopStickyContainer--companylocation').text
        except:
            #location='None'
            location=location=each.find('div', class_='icl-u-xs-mt--xs').find_next('a').text

    for each in soup.find_all(class_="jobsearch-CompanyInfoWithoutHeaderImage"):
        try:
            company=each.find('div', class_='icl-u-xs-mt--xs').find_next('span').text
        except:
            company=each.find(class_='icl-u-xs-mt--xs').find_next('a').text
    
    
    for each in soup.find_all(class_="jobsearch-JobComponent-description"):
        try:
            description=each.find(class_='jobsearch-jobDescriptionText').text
        except:
            description='None'
        
        df=df.append({'Title':title, 'Salary':salary,'Location':location, 'Company':company, 'Description':description}, ignore_index=True)
    
    return df
```

# Scraping these 270 websites for Company and Job Description and put into 1 dataframe

```python
#trying a copy first....

Title=[]
Descriptions=[]
Company=[]
Location=[]

#tags=[]
#chunks=[]
for url in websites:
    response=get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    
    job_containers= soup.find('div', {"class":"jobsearch-jobDescriptionText"})
    company_containers=soup.find('div', class_='icl-u-xs-mt--xs').find_next('a')
    title_containers=soup.find('h3')
    location_containers=soup.find('div', class_='icl-u-xs-mt--xs').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div').find_next('div')
    
    job_descriptions=str(job_containers.text)
    job_company=str(company_containers.text)
    title_names=str(title_containers.text)
    locations=str(location_containers.text)
    
    tokenize_descriptions=nltk.word_tokenize(job_descriptions)
    tokenize_company=nltk.word_tokenize(job_company)
    tokenize_title=nltk.word_tokenize(title_names)
    tokenize_location=nltk.word_tokenize(locations)
    
    #tagging=nltk.pos_tag(tokenize_reviews)
    
    #chunk=nlp(job_descriptions)
    #spacy_NP=[chunk.text for chunk in chunk.noun_chunks]
    
    Company.append(tokenize_company)
    Descriptions.append(tokenize_descriptions)
    Title.append(tokenize_title)
    Location.append(tokenize_location)
    #tags.append(tagging)
    #chunks.append(spacy_NP)
```
